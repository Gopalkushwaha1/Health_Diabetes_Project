```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate sample data
np.random.seed(0)
data = np.random.exponential(scale=1, size=5000)

# Apply transformations
log_data = np.log(data)
reciprocal_data = 1 / data
sqrt_data = np.sqrt(data)

# Function to plot Q-Q plot
def plot_qq(data, title):
    plt.figure(figsize=(6, 6))
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f'Q-Q Plot: {title}')
    plt.show()

# Plotting Q-Q plots for original and transformed data
plot_qq(data, 'Original Data')
plot_qq(log_data, 'Logarithm Transformed Data')
plot_qq(reciprocal_data, 'Reciprocal Transformed Data')
plot_qq(sqrt_data, 'Square Root Transformed Data')
```
